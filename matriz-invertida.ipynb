## Funcion para hallar el determinante de una matriz 2x2

para calcular el valor determinante de una matriz de 2x2.
* -Se multiplican los indices iguales
* -Se multiplican los indices distintos
* -Se halla la diferencia restando la primera multiplicacion - la segunda multiplicacion

Cabe decir que si el determinante de una matriz es = a 0, entonces la matriz se la considera *singular* y no puede hallarse su matriz invertida

* -su único parametro es la matriz dada

```python
def determinante(matrix):
    if len(matrix) and len(matrix[0]) and len(matrix[1]) ==2 :
        x=(matrix[0][0]*matrix[1][1])-(matrix[0][1]*matrix[1][0])
        print(f"\nCalculamos la resta del producto de sus elementos cruzados \n    ({matrix[0][0]}x{matrix[1][1]}) - ({matrix[0][1]}x{matrix[1][0]}) = {x} \n {x} es el determinante de esta matriz 2x2")
        return x
    #print("Así sabremos que la determinante de la siguiente matriz:")
    #matext=[[print(l) for l in matrix]]
    else:
        return "La matrix no es 2x2"
```

## Funcion para eliminar filas y columnas (i,j) de una matriz dada

* Esta función la necesitaremos para hallar los determinantes de matrices superiores a 2x2
* iremos despejando filas y columnas en una función regresiva hasta quedarnos con matrices de 2x2 ;)

```python
def eli(m,f,c):
    print("\n--ENTRAMOS EN LA FUNCIÓN eli()")
    #sin_fila=[fil for i,fil in enumerate(m) if i != f]
   
    sin_fila=[m[i] for i in range(len(m)) if i != f]
    
    for i in range(len(sin_fila)):
        sin_fila[i]=[ele for j, ele in enumerate(sin_fila[i]) if j != c]
    print(f"\nMatriz eliminando indices ({f},{c}) :")
    for fila in sin_fila:
        print(fila)
    return sin_fila
```

## Funcion recursiva para hallar el determinante

Esta función devuelve el determinante de cualquier matriz n x n.
* - Si la matriz es 1x1 el determinante es el valor mismo de la matriz
* - Si la matriz es 2x2 el determinante se haya como explicamos anteriormente
* - Si la matriz es 3x3 o superior:
      *  1:Generaremos submatrices con la funcion eli(m,f,c) eliminando determinadas filas y columnas de la matriz dada, para hallar el determinante de estas y generar una nueva matriz
      *  2:Emplearemos el cofactor para saber si cambia el signo (+ o -) o no. Paa esto sumaremos los indices de la fila y la columna correspondiente y si la suma es par , no es necesario cambiar el signo. si la suma es impar el signo sí cambiará.
      * 3: incorporamos cada fila con los determinantes y con su cofactor aplicado en una nueva matriz      

```python
def recu(matriz):
    print("\n--ENTRAMOS DE NUEVO EN LA FUNCIÓN recu()--")    
    n =len(matriz)
    if n == 1:  
        return matriz[0][0]
    elif n == 2: 
        print("    -como la sub-matriz resultante es 2x2")
        print("\n--ENTRAMOS EN LA FUNCION DETERMINANTE--")
        return determinante(matriz)
    
    else :
        print("\n\nPara esto, necesitamos una matriz 2x2, asi que eliminamos filas y columnas de manera \
ordenada hasta que nos quede una matriz 2x2. \n\nAqui os mostraremos como:")
        det =0
        kuk=[]
        for j in range(n):
            submatriz=eli(matriz,0,j)
            #print(f"\nEliminamos la fila 1 y la columna {j+1} y llamamos a nuestra función para calcular la determinante de la siguiente matriz:\n")
            #matext=[[print(l) for l in submatriz]]
            cofactor = matriz[0][j] * ((-1) ** j)
            det+= cofactor * recu(submatriz) 
            temp = det if not kuk  else det- sum(kuk) 
            kuk.append(temp)
        print("     ----------------------------    ")
        print("\nAplicando la formula de los determinantes obtenemos:\n")
        for i in range(len(kuk)):
            if i == len(kuk)-1:
                print("(",matriz[0][i], "x (",kuk[i], ")) = ", det)
            elif i % 2 != 0:
                print("(",matriz[0][i],"x (",kuk[i], ")) - ", end="")
            else:
                print("(",matriz[0][i],"x (",kuk[i], ")) + ", end="")
        print("\n     ----------------------------     ")
        print(f"{det} es el determinante de la matriz dada.\n\nNos guardaremos el número para el último paso")
        return det
```

## Funcion transponer

* -Trasponemos las filas por las columnas
* -Usamos un 2 bucles for anidados simples para trasponer lo que eran filas por columnas y viceversa
* -La matriz transpuesta es necesaria para calcular la matriz inversa ya que hay que transponer la matriz adjunta que será nuestro siguiente paso

```python

def transpu(m):
    print("\nTenemos que transponer la matriz que nos has dado")
    print("------------------------->")
    print("\n--ENTRAMOS EN LA FUNCIÓN transpu--")
    print("\nY basicamente lo que eran filas ahora son columnas")
    print("\nMatriz original")
    for fila in m:
        print(fila)
    new=[]
    for i in range(len(m[0])):
        row=[]
        for j in range(len(m)):
            row.append(m[j][i])
        new.append(row)
    print(f"\nMatriz transpuesta")
    for fila in new:
        print(fila)
    return new
```

## Funcion Adjunta

* -La matriz adjunta se conforma de determinantes atendiendo a su cofactor en función de la posición en la matriz
* -Estos determinantes ya se explicaron anteriormente como se hallaban
* Respecto al codigo:
 * prestar atención a i y j en el bucle, que como valor inicial tinen 0, pero las matrices no tienen indice (0,0), por eso sumamos 1.
* Tambié debemos fijarnos en *if ((i+1)+(j+1)) % 2 !=0:* ya que si no englobamos con () toda la operación que queremos obtener %

```python
def adjunta(m):
    print("\nAhora crearemos la matriz adjunta de esa matriz traspuesta")
    print("------------------------->")
    print("--ENTRAMOS EN LA FUNCION adjunta()")
    print("\nEn esta parte también eliminaremos filas y columnas para hallar el determinante, pero esta vez le aplicaremos el co-factor")
    print("\nSobre nuestra matriz transpuesta te mostramos que \nfilas y columnas eliminamos y la matriz resultante")
    print("\nMatriz Transpuesta")
    for fila in m:
        print(fila)
    new=[]
    for i in range(len(m[0])):
        row=[]
        for j in range(len(m)):
            x=recu(eli(m,i,j))
            print(f"\nHallamos su determinante: {x}")
            #print(x, "at indice matriz",i+1,j+1,"y su suma es ",(i+1+j+1))
            if ((i+1)+(j+1)) % 2 !=0: 
                (print(f"Cofactor -1 al ser IMPAR la suma de sus indices \nFila {i+1}, Columna {j+1}={i+j+2}"))
                x*=-1
                row.append(x)
            else:
                row.append(x)
                print(f"Cofactor de 1 Al ser PAR la suma de sus indices \nFila {i+1}, Columna {j+1} = {j+i+2}")
        new.append(row)
    print("-------------------------")
    print("\nPor tanto de esta matriz transpuesta \nMatriz transpuesta")
    for fila in m:
        print(fila)
    print("\nObtenemos la Matriz adjunta")
    for fila in new:
        print(fila)
    print("\n-------------------------")
    print("\n\nComo último paso dividiremos cada elemento de esta matriz por el determinante de la matriz original dada")
    return new      
```

# funcion inversa

* -Finalmente crearemos una función que empleará todas las anteriores ya sea directa o indirectamente.
* La matriz inversa de una matriz es igual a **la matriz adjunta de su matriz traspuesta, dividida por su determinante**, siempre que este no sea 0
* Eso es lo que haremos:
   * un primer condicional para comprobar si la matriz no tiene inversa, para esto hacemos el if , para ver si el determinante de la matriz que queremos hallar su inversa es 0. Si es 0 no tiene inversa.
   * Si es distinto de 0, realizaremos lo que indica la definición

```python
def inversa(m):
    print("\nPrimero comprobemos si su determinante es 0 o no")
    print("------------------------->")
    print("--ENTRAMOS EN LA FUNCIÓN recu()--")
    print("\nAl ser recursiva entraremos hasta que los bucles calculen todas las matrices 2x2")
    det = recu(m)
    if det == 0:
        print("La matriz es 'singular'")
        return None  # Devolver None si la matriz es singular
    else:
        print("\nComo el resultado del determinante no es 0 sabemos que SI se puede invertir \n\nContinuemos con el siguiente paso")
        adj = adjunta(transpu(m))
        print(f"\n\nY como último paso dividimos cada elemento de la matriz adjunta por el determinante calculado de la matriz dada que era {det}")
        print("------------------>")
        print("\nEsto nos dara los resultados finales de nuestra matriz original ya invertida")
        #inv = [[elem / det for elem in fila] for fila in adj]
        inv=[]
        for fila in adj:
            row=[]
            for elem in fila:
                print(f"\nDividimos {elem} entre nuesro determinante calculado {det} \n\nSiendo:{elem}/{det} = {elem/det}")
                row.append(round(elem/det, 1))
            inv.append(row)
        print("\n Por tanto, al colocar nuestros resultados en sus correspondientes índices obtenemos nuestra matriz invertida:")
        print("\n----MATRIZ INVERTIDA----")
        for fila in inv:
            print(fila)
        print("\n\nSiendo esta la matriz dada")
        print("\n----MATRIZ ORIGINAL----")
        for fila in m:
            print(fila)
        return inv
```
