```python
def gradient_descent( x0, f, paso, lim):
    """Esta función recive:
    x0: un punto inicial
    f: una función
    paso: una longitud de paso
    lim_iter: (opcional) un limite maximo de iteraciones"""
    n=0
    inicial = x0
    while n< lim and f(inicial) > 0.0000001:
        inicial = inicial-(paso*der(f,inicial, h=0.0001))
        n+=1
        if n%50==0:
            print("iteración",n,inicial)
    return inicial

gradient_descent(x0,funtes,paso, 1200)
```

```python
#input para dos dimensiones
x0= 2
y0= 1
def fun2d(x,y):
    return 3*x - (-2)*x**3 +2*y**3
paso= 0.0001
h= 0.0001
limi= 2000
```

```python
# ahora con dos dimensiones vector x,y
def gradi2d(x0,y0,f,paso,lim ):
    x=x0
    y=y0
    n=0
    while n < lim and f(x,y) > 0.000001:
        #calculamos derivadas parciales
        parciX = (f(x+h,y)-f(x,y))/(2*h)
        parciY = (f(x, y +h)-f(x,y))/(2*h)
        # actualizamos nuestros x e y
        x = x - (paso * parciX)
        y = y - (paso * parciY)
        n+=1
        if n %50 ==0:
            print("Iteracion ", n, "en x,y =","(", round(x,2),",",round(y,2),")")
    return x,y
gradi2d(x0,y0,fun2d, paso, limi)
```
