## Function to find the determinant of a 2x2 matrix.

To calculate the determinant of a 2x2 matrix:
* -Multiply the like indices
* -Multiply the different indices
* -Find the difference by subtracting the second multiplication from the first multiplication

It should be noted that if the determinant of a matrix equals 0, then the matrix is considered *singular* and its inverse matrix cannot be found.

* -Its only parameter is the given matrix

```python
def determinant(matrix):
    if len(matrix) and len(matrix[0]) and len(matrix[1]) ==2 :
        x=(matrix[0][0]*matrix[1][1])-(matrix[0][1]*matrix[1][0])
        print(f"\nWe calculate the subtraction of the product of its cross elements. \n    ({matrix[0][0]}x{matrix[1][1]}) - ({matrix[0][1]}x{matrix[1][0]}) = {x} \n {x} is the determinat of this 2x2 matrix")
        return x
    else:
        return "The matrix is not 2x2"
```

## Function to eliminate rows and columns (i, j) from a given matrix.

* We will need this function to find the determinants of matrices larger than 2x2.
* We will progressively eliminate rows and columns in a recursive function until we are left with 2x2 matrices ;)


```python
def delete(m,f,c):
    print("\n--WE ENTER THE FUNCTION delete()")
    sin_fila=[m[i] for i in range(len(m)) if i != f]
    
    for i in range(len(sin_fila)):
        sin_fila[i]=[ele for j, ele in enumerate(sin_fila[i]) if j != c]
    print(f"\nMatrix deleting index ({f},{c}) :")
    for fila in sin_fila:
        print(fila)
    return sin_fila
```

## Recursive function to find the determinant

This function returns the determinant of any n x n matrix.
* - If the matrix is 1x1, the determinant is the value of the matrix itself.
* - If the matrix is 2x2, the determinant is found as explained earlier.
* - If the matrix is 3x3 or larger:
      * 1: We generate submatrices using the eli(m, f, c) function by eliminating specific rows and columns from the given matrix to find the determinant of these and generate a new matrix.
      * 2: We use the cofactor to determine whether the sign (+ or -) changes or not. To do this, we sum the indices of the corresponding row and column, and if the sum is even, the sign remains unchanged; if the sum is odd, the sign changes.
      * 3: We incorporate each row with its determinant and its cofactor applied into a new matrix.
  

```python
def recursive(matriz):
    print("\n--WE ENTER THE FUNCTION recursive() AGAIN--")    
    n =len(matriz)
    if n == 1:  
        return matriz[0][0]
    elif n == 2: 
        print("    -Asi the result sub-matrix is 2x2")
        print("\n--WE ENTER IN THE determinat() FUNCTION--")
        return determinant(matriz)
    
    else :
        print("\n\nFor this, we need a 2x2 matrix, so we eliminate rows and columns in a \
sequential way until we are left with a 2x2 matrix. \n\nHere we will show you how:")
        det =0
        kuk=[]
        for j in range(n):
            submatriz=delete(matriz,0,j)
            cofactor = matriz[0][j] * ((-1) ** j)
            det+= cofactor * recursive(submatriz) 
            temp = det if not kuk  else det- sum(kuk) 
            kuk.append(temp)
        print("     ----------------------------    ")
        print("\nApplying the determinant formula, we obtain:\n")
        for i in range(len(kuk)):
            if i == len(kuk)-1:
                print("(",matriz[0][i], "x (",kuk[i], ")) = ", det)
            elif i % 2 != 0:
                print("(",matriz[0][i],"x (",kuk[i], ")) - ", end="")
            else:
                print("(",matriz[0][i],"x (",kuk[i], ")) + ", end="")
        print("\n     ----------------------------     ")
        print(f"{det} is the determinat for the given matrix.\n\nWe keep the number for the last step")
        return det
```

## transponse function

* -We transpose rows with columns
* -We use two simple nested for loops to transpose rows into columns and vice versa
* -The transposed matrix is necessary to calculate the inverse matrix because we need to transpose the adjugate matrix, which will be our next step


```python

def transpose(m):
    print("\nWe have to transpose the matrix you gave")
    print("------------------------->")
    print("\n--WE ENTER TE transpose() FUNCTION--")
    print("\nAnd basically what were rows are now columns.")
    print("\nOriginal Matrix")
    for fila in m:
        print(fila)
    new=[]
    for i in range(len(m[0])):
        row=[]
        for j in range(len(m)):
            row.append(m[j][i])
        new.append(row)
    print(f"\nTranspose Matrix")
    for fila in new:
        print(fila)
    return new
```

## Adjugate Function

* The adjugate matrix consists of determinants considering their cofactor based on their position in the matrix.
* These determinants were previously explained on how to find them.
* Regarding the code:
  * Pay attention to i and j in the loop, which start with 0 as initial values, but matrices don't have an index (0,0), hence we add 1.
  * Also, we need to pay attention to *if ((i+1)+(j+1)) % 2 != 0:*, as if we don't encapsulate the entire operation we want to obtain with ()


```python
def adjugate(m):
    print("\nNow we create the adjugate maxtix from the transpose one")
    print("------------------------->")
    print("--WE ENTER IN THE adjugate() FUNCTION")
    print("\nNow we also eliminate rows and coloumns to find the determinante, but this time we to use the co-factor")
    print("\nSFrom out transpose matrix we show you the \neliminatexd rows and columns and the result matrix")
    print("\Transpose Matrix")
    for fila in m:
        print(fila)
    new=[]
    for i in range(len(m[0])):
        row=[]
        for j in range(len(m)):
            x=recursive(delete(m,i,j))
            print(f"\nWe find his determinant: {x}")
            if ((i+1)+(j+1)) % 2 !=0: 
                (print(f"The co-factor is -1 when the sum of his indeces is EVEN \nRow {i+1}, Column {j+1}={i+j+2}"))
                x*=-1
                row.append(x)
            else:
                row.append(x)
                print(f"The co-factor is 1 when the sum of his indeces is ODD \nRiw {i+1}, Column {j+1} = {j+i+2}")
        new.append(row)
    print("-------------------------")
    print("\nSo, for this transpose matrix \nTranspose Matrix")
    for fila in m:
        print(fila)
    print("\nWe got the adjugate Matrix")
    for fila in new:
        print(fila)
    print("\n-------------------------")
    print("\n\nAs a final step, we will divide each element of this matrix by the determinant of the original given matrix.")
    return new      
```

# Inverse function

* Finally, we will create a function that will use all the previous functions either directly or indirectly.
* The inverse matrix of a matrix is equal to **the adjugate matrix of its transposed matrix, divided by its determinant**, provided that the determinant is not 0.
* This is what we will do:
  * First, a conditional to check if the matrix has no inverse, using an if statement to check if the determinant of the matrix we want to invert is 0. If it is 0, it has no inverse.
  * If it is not 0, we will proceed with what the definition indicates.

```python
def inverse(m):
    print("\nFirst check if its determinat is 0 or not")
    print("------------------------->")
    print("--WE ENTER IN THE recursive() FUNCTION--")
    print("\nWe go into recusrsive function for each 2x2 matrix in the matrix")
    det = recursive(m)
    if det == 0:
        print("'Singular' Matrix")
        return None  # return None if the matrix is singular
    else:
        print("\nSince the determinant result is not 0, we know it CAN be inverted.\n\nLet's continue with the next step.")
        adj = adjugate(transpose(m))
        print(f"\n\nAnd as a final step, we divide each element of the adjugate matrix by the calculated determinant of the given matrix, which was {det}.")
        print("------------------>")
        print("\nEhis will give us the final results of our original inverted matrix.")
        inv=[]
        for fila in adj:
            row=[]
            for elem in fila:
                print(f"\nDividing {elem} by our calculated determinant {det}\n\nResulting in: {elem}/{det} = {elem/det}")
                row.append(round(elem/det, 1))
            inv.append(row)
        print("\n Therefore, placing our results in their corresponding indices gives us our inverted matrix:")
        print("\n----INVERTED MATRIX----")
        for fila in inv:
            print(fila)
        print("\n\nThis is the given matrix")
        print("\n----ORIGINAL MATRIX----")
        for fila in m:
            print(fila)
        return inv
```
